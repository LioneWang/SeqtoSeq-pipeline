```python
import random
all_indexes = set([i for i in range(110380)])
indexes = list(set(random.choices([i for i in range(110380)], k=20000)))
test_indexes = set(indexes[:10000])
dev_indexes = set(indexes[10000:])
train_indexes = all_indexes - test_indexes - dev_indexes
```

```python
def save_dataset(name_en, name_zh, indexes, list_en, list_zh):
    res_en, res_zh = "", ""
    count = 0
    c2 = 0

    for index in indexes:
        try:
            c2 += 1
            res_en += list_en[index] + "\n"
            res_zh += list_zh[index] + "\n"
        except:
            count += 1
    print(count, c2)

    open(name_en, "w").write(res_en)
    open(name_zh, "w").write(res_zh)
```

```python
save_dataset("dev.en", "dev.zh", dev_indexes, list_en, list_zh)
save_dataset("test.en", "test.zh", test_indexes, list_en, list_zh)
save_dataset("train.en", "train.zh", train_indexes, list_en, list_zh)
```
